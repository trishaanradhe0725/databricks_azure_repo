```python
server_name=dbutils.secrets.get('kv-practice','Sqlserverconnection')
database_name=dbutils.secrets.get('kv-practice','sqldatabase')
user_name=dbutils.secrets.get('kv-practice','username')
pwd=dbutils.secrets.get('kv-practice','password')
```

```python
table_name = "[dbo].[Employees]"

df = spark.read \
    .format("jdbc") \
    .option("url", f"jdbc:sqlserver://{server_name}:1433")\
    .option("database", database_name) \
    .option("dbtable", table_name) \
    .option("user", user_name) \
    .option("password", pwd) \
    .option("driver", "com.microsoft.sqlserver.jdbc.SQLServerDriver") \
    .option("encrypt", "true") \
    .load()

df.show(10)
```

```python
df.write \
    .format("jdbc") \
    .option("url", f"jdbc:sqlserver://{server_name}:1433;database={database_name}") \
    .option("dbtable", table_name) \
    .option("user", user_name) \
    .option("password", pwd) \
    .option("driver", "com.microsoft.sqlserver.jdbc.SQLServerDriver") \
    .option("encrypt", "true") \
    .mode("overwrite")\
    .save()
```
