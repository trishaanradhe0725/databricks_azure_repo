```python
# At the very top of your notebook
dbutils.widgets.dropdown("environment", "dev", ["dev", "test", "prod", "uat"])

# Then later...
env = dbutils.widgets.get("environment").strip().lower()

SCOPE_MAPPING = {
    "dev":  "kv-dev",
    "test": "kv-test",
    "uat":  "kv-uat",
    "prod": "kv-prod"
}

secret_scope = SCOPE_MAPPING.get(env)

if secret_scope is None:
    raise ValueError(f"Unknown environment: {env}. Supported: {list(SCOPE_MAPPING.keys())}")

print(f"Using secret scope: {secret_scope}")
```
