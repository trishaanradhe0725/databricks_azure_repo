```python
from pyspark.sql.types import StructField,StringType,StructType,IntegerType
from pyspark.sql.functions import col
```

```python
data=[(1,'murali',2000,1),(2,'krish',3000,2),(3,'hari',4000,3)]
schema = StructType([
                    StructField('ID',IntegerType()),
                    StructField('Name',StringType()),
                    StructField('Salary',IntegerType()),
                    StructField('Dept',IntegerType())
                    ])
emp_df=spark.createDataFrame(data,schema)
emp_df.display()                    
```

```python
dept_data=[(1,'IT'),(2,'HR'),(3,'FINANCE')]
dept_schema = StructType([
StructField('dept_id',IntegerType()),
StructField('dept_name',StringType())
])
dept_df=spark.createDataFrame(dept_data,dept_schema)
dept_df.display()
```

```python
emp_dpt_df=emp_df.join(dept_df,col('Dept')==col('dept_id'),'inner')
emp_dpt_df.display()
```

```python
count=emp_dpt_df.filter(col("dept_name")==_dname).count()
```

```python
if count>=1:
    print("Records are present")
else:
    print("No Records are present")
```
